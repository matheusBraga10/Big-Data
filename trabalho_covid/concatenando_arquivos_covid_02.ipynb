**Inicializando a plataforma**

```python
lista_registros_covid_19 = []
lista_nome_arquivo = []
diretorio_covid_path_matheus = "/content/drive/MyDrive/ANÁLISE E DESENVOLVIMENTO DE SISTEMAS/4⁰ Período /Bid-Data/Trabalho DATA-TOWN/trabalho_final/database/HIST_PAINEL_COVIDBR_29set2023"

for arquivo in os.listdir(diretorio_covid_path_matheus, ):
  if arquivo.endswith('.csv'):
    nome_arquivo = '/' + arquivo
    nome_registro_covid = pd.read_csv(diretorio_covid_path_matheus + nome_arquivo, sep=';', encoding='latin-1')
    print('Nome do arquivo: ', arquivo)
    lista_registros_covid_19.append(nome_registro_covid)

```

- Criando uma lista de todos os arquivos, para concater em um só arquivo

```python
contador = 1
lista_registros_covid_concatenado = lista_registros_covid_19[0]
for i in lista_registros_covid_19[1:]:
  lista_registros_covid_concatenado = pd.concat([lista_registros_covid_concatenado, lista_registros_covid_19[contador]])
  contador = contador + 1
```

- Escrevendo e salvando os arquivos concatenados em um só arquivo


```python
path_matheus_concatenado = '/content/drive/MyDrive/ANÁLISE E DESENVOLVIMENTO DE SISTEMAS/4⁰ Período /Bid-Data/Trabalho DATA-TOWN/trabalho_final/database/arquivo_tratado/arquivo_concatenado_covid.csv'
lista_registros_covid_concatenado.to_csv(path_matheus_concatenado, index=False)
```
