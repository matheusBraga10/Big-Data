- Carregando arquivo de Vacinação por data, Covid Brasil

```python
path_matheus_vacina_data = '/content/drive/MyDrive/ANÁLISE E DESENVOLVIMENTO DE SISTEMAS/4⁰ Período /Bid-Data/Trabalho DATA-TOWN/trabalho_final/database/DADOS_VACINACAO_COVID-19/vacina_por_data.xlsx'
df_vacina_data = pd.read_excel(path_matheus_vacina_data)
```

- Vizualisando arquivo vacina

```python
df_vacina_data
```

- Alterando nome da coluna para facilitar o Join

```python
df_vacina_data.rename(columns={'Data da Vacina' : 'data'}, inplace=True)
df_vacina_data
```

- Unificando o formato e tipo de data para Join

```python
df_vacina_data['data'] = pd.to_datetime(df_vacina_data['data'])

df_vacina_data
```

```python
df_brasil['data'] = pd.to_datetime(df_brasil['data'])

df_brasil
```

- Unindo colunas pelo método Merge Join do pandas

```python
tabela_geral_covid_vacina_brasil = df_brasil.merge(df_vacina_data, how='outer', on='data')
tabela_geral_covid_vacina_brasil.shape
```

```python
inicio2 = pd.to_datetime(tabela_geral_covid_vacina_brasil['data']).dt.date.min()
fim2 = pd.to_datetime(tabela_geral_covid_vacina_brasil['data']).dt.date.max()
print(inicio2)
print(fim2)
```

- Entendendo esses valores

- Excluindo esses valores

```python
tabela_geral_covid_vacina_brasil = tabela_geral_covid_vacina_brasil[filtro]
```

- Excluindo colunas vazias

```python
del tabela_geral_covid_vacina_brasil['municipio']
del tabela_geral_covid_vacina_brasil['estado']
del tabela_geral_covid_vacina_brasil['codmun']
del tabela_geral_covid_vacina_brasil['codRegiaoSaude']
del tabela_geral_covid_vacina_brasil['nomeRegiaoSaude']
del tabela_geral_covid_vacina_brasil['interior/metropolitana']

tabela_geral_covid_vacina_brasil.isna().sum()
```
